### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values and corresponding y values (square of x)
x = np.linspace(0, 10, 100)
y = x**2

# Create the plot
plt.plot(x, y)

# Customize the plot with title, labels, and grid
plt.title('Line Plot of x vs x^2')
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.grid(True)

# Display the plot
plt.show()
```
